```python
df = pd.read_csv('./data/prophet_ds.csv')
```

```python
pf = Prophet(seasonality_mode='additive')
pf.fit(df)
forecast = pf.predict(pf.make_future_dataframe(34, freq='D'))
fig = pf.plot(forecast)
```
